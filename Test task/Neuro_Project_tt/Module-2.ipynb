```python
rich_surv_data = surveys_data.loc[(surveys_data['Ваш уровень дохода'] == '75.000 – 100.000 рублей') | (surveys_data['Ваш уровень дохода'] == 'более 100.000 рублей')]
rich_surv_data      # создали группу обеспеченных
```

```python
aver_surv_data = surveys_data.loc[(surveys_data['Ваш уровень дохода'] == '50.000 – 75.000 рублей') | (surveys_data['Ваш уровень дохода'] == '30.000 – 50.000 рублей')]
aver_surv_data      # создали группу среднего заработка
```

```python
need_surv_data = surveys_data.loc[(surveys_data['Ваш уровень дохода'] == '15.000 – 30.000 рублей') | (surveys_data['Ваш уровень дохода'] == 'до 15.000 рублей')]
need_surv_data      # низкий уровень заработка 
```

Разбили на три группы по доходу, для следющего анализа. Проверили количество значений в каждой таблице, для того, что бы сверить не потеряны ли данные. 
Уже заметно, что у респондентов с самым большим доходом, более определенные описания главного героя, такие как: "Пловец и гимнастка".

```python
rich_list = rich_surv_data['cid '].to_list()
aver_list = aver_surv_data['cid '].to_list()
need_list = need_surv_data['cid '].to_list()
# отобрали пользователей по нашим группам
```

```python
rich_emotions_df = emotions_data[emotions_data.cid.isin(rich_list)]
aver_emotions_df = emotions_data[emotions_data.cid.isin(aver_list)]
need_emotions_df = emotions_data[emotions_data.cid.isin(need_list)]
all_emotions = ['happy', 'angry', 'fear', 'disgusted', 'sad', 'surprised', 'calm', 'confused']
all_emot_dict = {'happy': 'mean', 'angry': 'mean', 'fear': 'mean', 'disgusted': 'mean', 
'sad': 'mean', 'surprised': 'mean', 'calm': 'mean', 'confused': 'mean'}
```

```python
rich_gr = rich_emotions_df.groupby('episode', as_index=False).agg(all_emot_dict).reset_index()
aver_gr = aver_emotions_df.groupby('episode', as_index=False).agg(all_emot_dict).reset_index()
need_gr = need_emotions_df.groupby('episode', as_index=False).agg(all_emot_dict).reset_index()
```

```python
rich_gr.plot(x='episode', y=all_emotions, figsize=(20, 10))
```

```python
aver_gr.plot(x='episode', y=all_emotions, figsize=(20, 10))
```

```python
need_gr.plot(x='episode', y=all_emotions, figsize=(20, 10))
```

```python
higher_education_data = surveys_data.loc[(surveys_data['Ваш род деятельности'] == 'Специалист с высшим образованием')]
higher_education_list = higher_education_data['cid '].to_list()
higher_education_df = emotions_data[emotions_data.cid.isin(higher_education_list)]
higher_education_gr = higher_education_df.groupby('episode', as_index=False).agg(all_emot_dict).reset_index()
higher_education_gr.plot(x='episode', y=all_emotions, figsize=(20, 10))
```

```python
lot_of_emotions_dict = {}
for i in all_emotions:
    temp = emotions_data.query(f"{i} > 90").cid.mode()
    max_numb = 90
    while len(temp) > 1 and max_numb <= 100:
        max_numb =+ 1
        temp = emotions_data.query(f"{i} > {max_numb}").cid.mode()
    lot_of_emotions_dict[i] = [temp]

lot_of_emotions_df = pd.DataFrame(list(lot_of_emotions_dict.items()), columns=['emotions', 'cid'])
lot_of_emotions_df 
```
